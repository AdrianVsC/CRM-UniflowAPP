```python
# se importan las librerias necesarias 
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os 
from supabase import create_client, Client
from dotenv import load_dotenv
```

```python
# Cargar variables de entorno
load_dotenv()

# Obtener variables de entorno
url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)

# Verificar que las variables no sean None
if url is None or key is None:
    raise ValueError("SUPABASE_URL o SUPABASE_KEY no están configuradas en el archivo .env")

# Crear cliente de Supabase con manejo de excepciones
try:
    supabase: Client = create_client(url, key)
    print("Conexión a Supabase establecida exitosamente.")
except Exception as e:
    print(f"Error al conectar con Supabase: {e}")
    raise

def extraer_datos(tabla: str) -> pd.DataFrame:
    """
    Extrae datos de una tabla específica en Supabase y los convierte en un DataFrame.
    
    :param tabla: Nombre de la tabla en Supabase
    :return: DataFrame con los datos extraídos
    """
    response = supabase.table(tabla).select("*").execute()
    return pd.DataFrame(response.data)

# Extraer datos de las tres tablas necesarias para las visualizaciones
customer_df = extraer_datos("customers")
team_df = extraer_datos("teams")
condominium_df = extraer_datos("condominums")
```
